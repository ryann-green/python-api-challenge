```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import random as rd
from pprint import pprint 
from citipy import citipy
from api_key import weather_api_key
```

```python
#parameters
city=[]
api_key=weather_api_key

#begin an empty counter to hold how many successul results we've had
counter=0
```

```python
#begin empty lists
cities=[]
lat=[]
lng=[]
temp=[]
humidity=[]
cloudiness=[]
wind=[]
country=[]

#begin while loop
while counter<550:
    
    #find random latitudes and longitudes
    latitude=round(rd.randint(-90,89)+rd.random(),6)
    longitude=round(rd.randint(-180,179)+rd.random(),6)

    #find the nearest city of the lats & longs
    city=citipy.nearest_city(latitude,longitude)
    city_name=city.city_name
    
    #check to see if the city name is already in our cities list
    if city_name not in cities:
        
        #build our url
        url=f'http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}&units=Imperial'

        #get the url and read it as a json file
        response=requests.get(url).json()
        #pprint(response)
        
        #append our responses to our lists
        try:        
            temp.append(response['main']['temp'])  
            humidity.append(response['main']['humidity'])
            cloudiness.append(response['clouds']['all'])
            wind.append(response['wind']['speed'])
            lat.append(response['coord']['lat'])
            lng.append(response['coord']['lon'])
            country.append(response['sys']['country'])
                    
            if city_name not in cities:
                cities.append(city_name)
                counter+=1
                print(f"City {city_name} found, appending results as city number {counter}") 
        except KeyError: 
            print(f"{city_name} not found, skipping operation") 
            pass
            
            
    
```

```python
#create the dataframe and check the dataframe
df=pd.DataFrame({"Cities":cities,
              "Temp (F)":temp,
                "Humidity (%)":humidity,
                "Cloudiness (%)":cloudiness,
                "Wind Speed (mph)":wind,
                "Latitude":lat,
                "longitude":lng,
                "Country":country})

df.to_csv(r'city_data.csv',index=False)

df.head(10)
```

```python
#create individual dataframes for the lat & longs we need
temp_df=df.loc[:,["Latitude","Temp (F)"]]
hum_df=df.loc[:,["Latitude","Humidity (%)"]]
cloud_df=df.loc[:,["Latitude","Cloudiness (%)"]]
wind_df=df.loc[:,["Latitude","Wind Speed (mph)"]]
```

```python
 #create a function for linear regression and plot
def create_regr (regr):
    s_hem=regr.loc[df["Latitude"]<0,:]
    n_hem=regr.loc[df["Latitude"]>0,:]
    s_hem=s_hem.reset_index()
    n_hem=n_hem.reset_index()
    
    #locate the x values
    s_x=s_hem.iloc[:,1]
    n_x=n_hem.iloc[:,1]
    
    #get the column headers
    columnnames=list(s_hem.columns.values)
    y_label=columnnames[2]
    
    #locate the y values
    s_y=s_hem.iloc[:,2]
    n_y=n_hem.iloc[:,2]
    
    #find the s_hem correlation
    correlation = st.pearsonr(s_x,s_y)   
    
    #find the line regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(s_x,s_y)
    regress_values = s_x * slope + intercept
    
    #create a variable for the line equation
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    fig, [ax1, ax2] = plt.subplots(nrows=2,ncols=1,figsize=(8,8))
    
    #create scatter
    ax1.scatter(s_x,s_y,facecolors="darkgreen",edgecolors="black",alpha=.45)   
    ax1.set_title(f"Southern Hemisphere {y_label} vs Latitude",fontsize=14)    
    ax1.plot(s_x,regress_values,"r-",alpha=.6)
    ax1.set_ylabel(f"{y_label}",fontsize=13)
    ax1.annotate(line_eq,(min(s_x),max(s_y)-1),fontsize=13,color="red", alpha=1)
    print(f"The Southern Hemisphere's correlation  of {y_label} and Latitude is {correlation[0]}")
    
    #find the n_hem correlation
    correlation = st.pearsonr(n_x,n_y)   
    #find the line regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(n_x,n_y)
    regress_values = n_x * slope + intercept
    
    #create the variable for the line equation
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    #create scatter
    ax2.scatter(n_x,n_y,facecolors="mediumslateblue",edgecolors="black",alpha=.45) 
    ax2.set_title(f"Northern Hemisphere {y_label} vs Latitude",fontsize=14)    
    ax2.plot(n_x,regress_values,"r-",alpha=.6)
    ax2.set_ylabel(f"{y_label}",fontsize=13)
    ax2.annotate(line_eq,(min(n_x),min(n_y)),fontsize=13,color="red", alpha=1)
    
    print(f"The Northern Hemisphere's correlation of {y_label} and Latitude is {correlation[0]}")
    fig.savefig(f'plots/n_s_{y_label}_lat.png')
    
    return
 
```

## **Temperature vs Latitude Analysis**

```python
create_regr(temp_df)
```

- This linear regression model whos us that as the cities move closer to the equator from the poles, that the temperature increases (downward v shape when placed side by side)

```python
create_regr(hum_df)
```
